```python
from pypdf import PdfReader
```

```python
#Extracting Text from PDF
with open("/content/2312.11805v2.pdf", "rb") as f:
  pdf_reader = PdfReader(f)
  text = ''
  for i in range(pdf_reader.get_num_pages()):
    page_content = pdf_reader.pages[i]
    page_text = page_content.extract_text()
    text += page_text
    print(f'Text from page {i + 1}:\n{page_text}\n')
    print("-----------------------------------------------------------------")
```

```python
#Extracting Tables from PDF
import tabula
from tabula import read_pdf
tables = tabula.read_pdf("/content/2312.11805v2.pdf", pages="all")
```
